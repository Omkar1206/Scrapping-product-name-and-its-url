```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver =webdriver.Chrome('chromedriver.exe')
```

```python
url='https://www.consumerreports.org/cro/a-to-z-index/products/index.htm'
```

```python
driver.get(url)
```

```python
soup=BeautifulSoup(driver.page_source,'lxml')
```

```python
for div in soup.findAll('div',{'class':'col-lg-9 col-md-8 col-sm-12 col-xs-12'}):
    for x in div.findAll('a'):
        print('Name:',x.text)
        print('Url: https://www.consumerreports.org'+x.get('href'))
        print("*"*20)
```

```python
all_divs=soup.find_all('div',attrs={'class','entry-letter'})
```

```python
type(all_divs)
```

```python
for div in all_divs:
    print(div.div.a.span.string)
```
